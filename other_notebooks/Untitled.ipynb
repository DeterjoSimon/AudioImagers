```python
import os
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
```

```python
audio_fpath = 'data/'
x, sr = librosa.load(audio_fpath+'Vein drive cut.wav', sr=None)
print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
fig, ax = plt.subplots(figsize=(10,10), subplot_kw={'projection': 'polar'})
s = librosa.display.specshow(Xdb, ax=ax, sr=sr, x_axis='time', y_axis='hz')
ax.set_rticks([])
fig.colorbar(s, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,10), subplot_kw={'projection': 'polar'})
s = librosa.display.specshow(Xdb, ax=ax, sr=sr, x_axis='time', y_axis='log')
ax.set_rticks([])
fig.colorbar(s, ax=ax)
```
